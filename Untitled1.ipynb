```python
import numpy as np
import matplotlib.pyplot as pit
day=[1,2,3,4,5]
day=['april-15','april-22','june-3','june-16','march-10']
temp=[35,35,23,38,27]
wind=[28,21,22,18,21]
ypos=np.arange(len(day))
pit.bar(ypos-0.2,temp,color='g',width=0.4)
pit.bar(ypos+0.2,wind,color='r',width=0.4)
pit.xticks(ypos,day)
pit.xlabel("Days in week")
pit.ylabel("Temp and Wind---------->")
pit.title("Temperature and Wind For Next 10 Days")
pit.legend()
```
